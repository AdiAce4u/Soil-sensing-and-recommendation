```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("1_npk-temp-ph-crop.csv")
```

```python
df2=pd.read_csv("npk-ph-temp-humidty.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.drop(columns=["Soil_Type",	"Variety", "Crop", "Humidity","Rainfall"], inplace=True)
df2.drop(columns=["label","rainfall","humidity"],inplace = True)
```

```python
df2.columns=[ "Nitrogen",	"Phosphorus",	"Potassium",	"Temperature", 	"pH_Value"]
```

```python
df1.head()
```

```python
df2.tail()
```

```python
df=pd.concat([df1,df2])
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import RobustScale

# Distribution plots
df.hist(bins=30, figsize=(12, 8))
plt.tight_layout()
plt.show()

# Boxplots for outlier visual check
plt.figure(figsize=(12,6))
df.boxplot(column=df.columns)
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(
    n_estimators=200,
    contamination=0.1,  
    random_state=42
)

iso.fit(df)
df['anomaly'] = iso.predict(df)  
```

```python
print(df['anomaly'].value_counts())
print(df[df['anomaly'] == -1].head())
```
